```python
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np


```

```python
def visualize(path):
    df = pd.read_csv(filepath_or_buffer = path)
    dfmin = df.min(axis=0) ##taking minimum of each features
    dfmax = df.max(axis=0) ##taking maximum of each features
    dfavg = df.mean(axis=0) ##taking average of each features
    dfmin.plot(subplots=True, layout=(1,1)) ## creating plot with min values
    matplotlib.pyplot.show()
    dfmax.plot(subplots=True, layout=(1,1)) ## creating plot with max values
    matplotlib.pyplot.show()
    dfavg.plot(subplots=True, layout=(1,1)) ## creating plot with avg values
    matplotlib.pyplot.show()
```

```python
def identify_issues(path):
        df = pd.read_csv(filepath_or_buffer = path)
        dfmax = df.max()
        dfmin = df.min()
        if dfmax[0] == 0 and dfmax[20] == 20:
            print("I doubt the integrity of this dataset")
        else :
            if dfmin.sum() == 0:
                print("Why is this happening")
            else:
                print("Dataset looks good")
            
```

```python
dataset = sorted(glob.glob('data*.csv'))
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
    
```
